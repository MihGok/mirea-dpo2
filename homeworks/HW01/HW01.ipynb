Создание массивов разными методами

```python
py_list = [1, 2, 3, 4, 5]
arr_from_list = np.array(py_list)
print("Массив из списка:", arr_from_list)


zeros_arr = np.zeros((3, 3))
print("Массив нулей:\n", zeros_arr)


range_arr = np.arange(0, 10, 2)
print("Arange:", range_arr)
```

Демонстрация атрибутов массива

```python
test_arr = np.array([[1.5, 2.5, 3.5], [4.5, 5.5, 6.5]])
print("Форма массива (shape):", test_arr.shape)
print("Тип данных (dtype):", test_arr.dtype)
print("Размерность (ndim):", test_arr.ndim)
```

Срезы и доступ к элементам массива

```python
print("Первый элемент:", test_arr[0])
print("Последний элемент:", test_arr[-1])
print("Срез первого элемента:", test_arr[0][1:3])
print("Весь массив в обратном порядке:", test_arr[::-1])
```

Примеры векторных операций и базовых статистик

```python
print("Среднее для самого первого массива:", arr_from_list.mean())
print("Сумма первого и третьего массивов:", arr_from_list + range_arr)
print("Поементное произведение первого и третьего массивов:", arr_from_list * range_arr)
```
